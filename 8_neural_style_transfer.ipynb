#image processing fnctn

```python
def load_image(image,image_size=(512,512)):
  img=tf.io.read_file(image)
  img=tf.image.decode_image(img,channels=3)
  img=img=tf.image.convert_image_dtype(img,tf.float32)
  img=tf.image.resize(img,image_size,preserve_aspect_ratio=True)
  img=img[tf.newaxis,:]
  return img
```

```python
og_img=load_image("Original Image.png")
```

```python
og_img.shape
```

```python
style_img=load_image("Style Image.jpg")
```

#display function

```python
def imshow(image,title=None):
  if len(image.shape)>3:
    image = tf.squeeze(image,axis=0)
  plt.imshow(image)
  if title:
    plt.title(title)
```

#displaying image

```python
imshow(style_img)
```
